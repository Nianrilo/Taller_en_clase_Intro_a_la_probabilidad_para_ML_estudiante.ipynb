```python
from skimage.io import imread, imshow
from skimage.filters import prewitt_v,prewitt_h
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
Img1 = imread('citrus1.jpg')
Img2 = imread('citrus2.jpg')
Img3 = imread('leave1.png')
Img4 = imread('leave2.png')
plt.figure()
plt.imshow(Img1, cmap="gray")
plt.figure()
plt.imshow(Img2, cmap="gray")
plt.figure()
plt.imshow(Img3, cmap="gray")
plt.figure()
plt.imshow(Img4, cmap="gray")
```

```python
Ieg1 = cv2.cvtColor(Img1,cv2.COLOR_RGB2GRAY);
Ieg2 = cv2.cvtColor(Img2,cv2.COLOR_RGB2GRAY);
Ieg3 = cv2.cvtColor(Img3,cv2.COLOR_RGB2GRAY);
Ieg4 = cv2.cvtColor(Img4,cv2.COLOR_RGB2GRAY);
plt.figure()
plt.imshow(Ieg1, cmap="gray")
plt.figure()
plt.imshow(Ieg2, cmap="gray")
plt.figure()
plt.imshow(Ieg3, cmap="gray")
plt.figure()
plt.imshow(Ieg4, cmap="gray")
```

```python
ret1,mask1=cv2.threshold(Ieg1,100,100,cv2.THRESH_BINARY)
kernel1=np.ones((5,5),np.uint8)
mask1=cv2.erode(mask1,kernel1,iterations = 1)
ret2,mask2=cv2.threshold(Ieg2,100,100,cv2.THRESH_BINARY)
kernel2=np.ones((5,5),np.uint8)
mask2=cv2.erode(mask2,kernel2,iterations = 1)
ret3,mask3=cv2.threshold(Ieg3,100,100,cv2.THRESH_BINARY)
kernel3=np.ones((5,5),np.uint8)
mask3=cv2.erode(mask3,kernel3,iterations = 1)
ret4,mask4=cv2.threshold(Ieg4,100,100,cv2.THRESH_BINARY)
kernel4=np.ones((5,5),np.uint8)
mask4=cv2.erode(mask4,kernel4,iterations = 1)
plt.figure()
plt.imshow(mask1, cmap="gray")
plt.figure()
plt.imshow(mask2, cmap="gray")
plt.figure()
plt.imshow(mask3, cmap="gray")
plt.figure()
plt.imshow(mask4, cmap="gray")
```

```python
ret1,mask1=cv2.threshold(Img1,100,100,cv2.THRESH_BINARY)
kernel1=np.ones((5,5),np.uint8)
mask1=cv2.erode(mask1,kernel1,iterations = 1)
ret2,mask2=cv2.threshold(Img2,100,100,cv2.THRESH_BINARY)
kernel2=np.ones((5,5),np.uint8)
mask2=cv2.erode(mask2,kernel2,iterations = 1)
ret3,mask3=cv2.threshold(Img3,100,100,cv2.THRESH_BINARY)
kernel3=np.ones((5,5),np.uint8)
mask3=cv2.erode(mask3,kernel3,iterations = 1)
ret4,mask4=cv2.threshold(Img4,100,100,cv2.THRESH_BINARY)
kernel4=np.ones((5,5),np.uint8)
mask4=cv2.erode(mask4,kernel4,iterations = 1)
plt.figure()
plt.imshow(mask1, cmap="gray")
plt.figure()
plt.imshow(mask2, cmap="gray")
plt.figure()
plt.imshow(mask3, cmap="gray")
plt.figure()
plt.imshow(mask4, cmap="gray")
```

```python
edges_prewitt_vertical1 = prewitt_v(Ieg1)
edges_prewitt_vertical2 = prewitt_v(Ieg2)
edges_prewitt_vertical3 = prewitt_v(Ieg3)
edges_prewitt_vertical4 = prewitt_v(Ieg4)
plt.figure()
plt.imshow(edges_prewitt_vertical1, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_vertical2, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_vertical3, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_vertical4, cmap="gray")
```

```python
edges_prewitt_horizontal1 = prewitt_h(Ieg1)
edges_prewitt_horizontal2 = prewitt_h(Ieg2)
edges_prewitt_horizontal3 = prewitt_h(Ieg3)
edges_prewitt_horizontal4 = prewitt_h(Ieg4)
plt.figure()
plt.imshow(edges_prewitt_horizontal1, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_horizontal2, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_horizontal3, cmap="gray")
plt.figure()
plt.imshow(edges_prewitt_horizontal4, cmap="gray")
```

```python
img_blur1 = cv2.GaussianBlur(Ieg1,(5,5),0)
edges_canny1 = cv2.Canny(image=img_blur1,threshold1=50,threshold2=100)
img_blur2 = cv2.GaussianBlur(Ieg2,(5,5),0)
edges_canny2 = cv2.Canny(image=img_blur2,threshold1=50,threshold2=100)
img_blur3 = cv2.GaussianBlur(Ieg3,(5,5),0)
edges_canny3 = cv2.Canny(image=img_blur3,threshold1=50,threshold2=100)
img_blur4 = cv2.GaussianBlur(Ieg4,(5,5),0)
edges_canny4 = cv2.Canny(image=img_blur4,threshold1=50,threshold2=100)
plt.figure()
plt.imshow(edges_canny1, cmap="gray")
plt.figure()
plt.imshow(edges_canny2, cmap="gray")
plt.figure()
plt.imshow(edges_canny3, cmap="gray")
plt.figure()
plt.imshow(edges_canny4, cmap="gray")
```

```python
sum1 = np.sum(edges_canny1)
sum2 = np.sum(edges_canny2)
sum3 = np.sum(edges_canny3)
sum4 = np.sum(edges_canny4)
mean1 = np.mean(edges_canny1)
mean2 = np.mean(edges_canny2)
mean3 = np.mean(edges_canny3)
mean4 = np.mean(edges_canny4)
std1 = np.std(edges_canny1)
std2 = np.std(edges_canny2)
std3 = np.std(edges_canny3)
std4 = np.std(edges_canny4)
featuresi1 = [sum1,mean1,std1]
featuresi2 = [sum2,mean2,std2]
featuresi3 = [sum3,mean3,std3]
featuresi4 = [sum4,mean4,std4]
```
